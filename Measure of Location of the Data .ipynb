# percentile  - numpy library is use fully numerical calculation


```python
import numpy as np
```

```python
np.percentile(dataset['ssc_p'],10) #10% of ssc_p mark 
```

```python
np.percentile(dataset['ssc_p'],15) #  15% of ssc_p mark
```

```python
descriptive=pd.DataFrame(index=['Percentile:25%','Percentile:50%','Percentile:75%','Percentile:100%'],columns=quan)
for ColumnName in quan:
    descriptive[ColumnName]['Percentile:25%']=np.percentile(dataset[ColumnName],25)
    descriptive[ColumnName]['Percentile:50%']=np.percentile(dataset[ColumnName],50)
    descriptive[ColumnName]['Percentile:75%']=np.percentile(dataset[ColumnName],75)
    descriptive[ColumnName]['Percentile:100%']=np.percentile(dataset[ColumnName],100)
```

```python
descriptive
```

# Alternate Way percentile calculation in inbuilt function

```python
dataset.describe()
```

```python
alterwave=pd.DataFrame(index=['Q1:25%','Q2:50%','Q3:75%','Q4:100%'],columns=quan)
for ColumnName in quan:
    alterwave[ColumnName]['Q1:25%']=dataset.describe()[ColumnName]['25%']
    alterwave[ColumnName]['Q2:50%']=dataset.describe()[ColumnName]['50%']
    alterwave[ColumnName]['Q3:75%']=dataset.describe()[ColumnName]['75%']
    alterwave[ColumnName]['Q4:100%']=dataset.describe()[ColumnName]['max']
```
